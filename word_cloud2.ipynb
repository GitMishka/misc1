```python
sia = SentimentIntensityAnalyzer()

# Define target words
target_words = ["good", "staff", "time", "need", "improvement", "activities", "better"]

# Prepare a dataframe to store results
result_df = pd.DataFrame(columns=["Word", "Score", "Comment", "Overall Sentiment"])

# Iterate over each comment
for comment in df['Q24 - How can we improve in general?'].dropna():

    # Preprocessing text data
    # Converting the text data to lower case and removing any special characters
    comment = comment.lower().replace('[^\w\s]','')

    # Tokenize words
    words = word_tokenize(comment)

    # Determine overall sentiment of comment
    overall_sentiment_score = sia.polarity_scores(comment)
    transformed_overall_sentiment = (overall_sentiment_score['compound'] + 1) / 2 * 100  # This will map -1 to 0, 1 to 100, and 0 to 50

    # Check for target words
    for word in words:
        if word in target_words:
            # Analyze sentiment of word
            sentiment_score = sia.polarity_scores(word)
            # Transform compound score to range 0-100
            transformed_score = (sentiment_score['compound'] + 1) / 2 * 100  # This will map -1 to 0, 1 to 100, and 0 to 50
            # Append result to dataframe
            result_df = result_df.append({"Word": word, "Score": transformed_score, "Comment": comment, "Overall Sentiment": transformed_overall_sentiment}, ignore_index=True)
            # Break loop after finding first target word in comment
            break

# Export results to CSV
result_df.to_csv("sentiment_scores.csv", index=False)
```

```python
# Required Libraries
#!pip install nltk pandas
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.probability import FreqDist
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')

# Assuming you've loaded your data into a pandas dataframe named df
# df = pd.read_csv('path_to_your_data')

# Initialize sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Combine all the comments into one large string
text = ' '.join(comment for comment in df['Q24 - How can we improve in general?'].dropna())

# Tokenize words and remove stopwords
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
words = [word for word in words if word not in stop_words]

# Calculate word frequencies
freq_dist = FreqDist(words)

# Get 10 most common words
most_common_words = freq_dist.most_common(10)

# Prepare a dataframe to store results
result_df = pd.DataFrame(columns=["Word", "Frequency", "Score"])

# Analyze sentiment of most common words
for word, frequency in most_common_words:
    # Analyze sentiment of word
    sentiment_score = sia.polarity_scores(word)
    # Transform compound score to range 0-100
    transformed_score = (sentiment_score['compound'] + 1) / 2 * 100  # This will map -1 to 0, 1 to 100, and 0 to 50
    # Append result to dataframe
    result_df = result_df.append({"Word": word, "Frequency": frequency, "Score": transformed_score}, ignore_index=True)

# Display results
print(result_df)
result_df.to_csv("over_sentiment.csv", index=False)
```

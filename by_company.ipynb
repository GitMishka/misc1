```python
# Import necessary libraries
import pandas as pd
from google.colab import files
import io

```

```python
uploaded = files.upload()
azure_df = pd.read_csv(io.BytesIO(uploaded['Azure.csv']))
uploaded = files.upload()
map_df = pd.read_csv(io.BytesIO(uploaded['map.csv']))
# Upload Yammer.csv
uploaded = files.upload()
yammer_df = pd.read_csv(io.BytesIO(uploaded['Yammer.csv']))
```

```python
merged_df = pd.merge(azure_df, yammer_df, how='inner', left_on='userPrincipalName', right_on='User Principal Name')

# Merge the resulting dataframe with map_df on the userPrincipalName
merged_df = pd.merge(merged_df, map_df, how='left', on='userPrincipalName')

# Fill NaNs in the 'tab' column with 'no_match'
merged_df['tab'] = merged_df['tab'].fillna('!no_match')

# Create a summary dataframe for the total price for each tab
summary_df = merged_df.groupby('tab')['Price'].sum().reset_index()

# Save the dataframes to separate Excel worksheets
with pd.ExcelWriter('output.xlsx') as writer:
    # Save the summary dataframe as the first sheet
    summary_df.to_excel(writer, sheet_name='Summary', index=False)

    # Save each tab's data to separate sheets
    for tab, data in merged_df.groupby('tab'):
        # Excel doesn't allow sheet names longer than 31 characters
        if len(tab) > 31:
            tab = tab[:31]
        # Summarize price for current tab
        price_sum = data['Price'].sum()
        # Append a summary row to the current dataframe
        data = data.append({'Price': f'Total Price: {price_sum}'}, ignore_index=True)
        data.to_excel(writer, sheet_name=tab, index=False)

# Download the created file in your local system
files.download('output.xlsx')
```
